# Creating publication-ready game images

Using a combination of `pygambit` and the Gambit project's LaTeX graphics package `draw_tree`, we can generate publication quality images for games with just a few lines of code.

This tutorial will demonstrate the key functionality of `draw_tree` when used for games built with `pygambit`, using a "Kuhn poker" game derived from the Gambit catalog of games.
